```python
channels = ["北京二手台式机/配件","北京二手手机","北京二手笔记本"]
```

```python
def date_gen(date1,scope):
    the_time = date(date1.split('-')[0],date1.split('-')[1],date1.split('-')[2])
    add_time = timedelta(days=1)
    for i in range(int(scope)):
        yield the_time
        the_time += add_time
```

```python
def data_gen():
    for i in channels:
        data_list = []
        for j in date_gen('2016-1-1',7):
            data_list = list(json_db.find({pub_date:j},{'cates':i}))
            data_num = len(data_list)
        data = {
            'name': i,
            'data': data_num,
            'type': 'line',
            }
        yield data
```

```python
options = {
    'chart':{'zoomType':'xy'},
    'title':{'text':'发帖量统计'},
    'subtitle':{'text':'北京二手手机 北京二手笔记本 北京二手台式机/配件 最近7日内，发帖量变化情况'},
    'xAxis':{'categories':[j for j in date_gen('2016-1-1',7)]},
    'yAxis':{'title':{'text':'数量'}}
    }

series = [i for i in data_gen()]

charts.plot(series,show="inline",option=options)
```
