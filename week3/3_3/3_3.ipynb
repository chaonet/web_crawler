```python
channels = ["北京二手台式机/配件","北京二手手机","北京二手笔记本"]
```

```python
def date_gen(date1,scope):
    the_date = date(int(date1.split('-')[0]),int(date1.split('-')[1]),int(date1.split('-')[2]))
    add_date = timedelta(days=1)
    for i in range(int(scope)):
        yield the_date.strftime('%Y-%m-%d')
        the_date += add_date

for i in date_gen('2016-1-1',7):
    print i
```

```python
def data_gen(channels,date,scope):
    for i in channels:
        num_list = []
        for j in date_gen(date,scope):
            data_list = list(json_db.find({'pub_date':j,'cates':i}))
            data_num = len(data_list)
            num_list.append(data_num)
        data = {
            'name': i,
            'data': num_list,
            'type': 'line',
            }
        yield data
for i in data_gen(channels,'2016-1-1',7):
    print i
```

```python
options = {
    'chart':{'zoomType':'xy'},
    'title':{'text':'发帖量统计'},
    'subtitle':{'text':'北京二手手机 北京二手笔记本 北京二手台式机/配件 最近7日内，发帖量变化情况'},
    'xAxis':{'categories':[j for j in date_gen('2016-1-1',7)]},
    'yAxis':{'title':{'text':'数量'}}
    }

series = [i for i in data_gen(channels,'2016-1-1',7)]

charts.plot(series,show="inline",options=options)
```
