## 北京市各个城区，发帖量最多的3个大类目

```python
from pymongo import MongoClient
import charts

client = MongoClient('localhost', 27017, connect=False)
ganji = client['ganji']
json_db_new = ganji['json_db_new']
```

```python
area_list = []
for i in json_db_new.find().limit(300):
    area_list.append(i['area'][0])
areas = list(set(area_list))
print areas
print len(areas)
```

```python
def data_gen(area):
    pipeline = [
        {'$match':{'area':area}},
        {'$group':{'_id':{'$slice':['$cates',2,1]},'counts':{'$sum':1}}},
        {'$sort':{'counts':-1}},
        {'$limit':3}
    ]
    for i in json_db_new.aggregate(pipeline):
        data = {
            'name': i['_id'][0],
            'data': [i['counts']],
            'type': 'column',
            }
        yield data
for i in data_gen('朝阳'):
    print i
```

```python
series = [data for data in data_gen('朝阳')]
print series
```

```python
charts.plot(series,show="inline",options=dict(title=dict(text="朝阳区发帖量最多的3个大类目")))
```
